Hereâ€™s Python code to perform image translation using OpenCV. Translation involves shifting an image along the X and Y axes.

1. Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

2. Translate Image Function

```python
def translate_image(image, x_shift, y_shift):
    """
    Translates an image by shifting it along the X and Y axes.

    Args:
        image: Input image (NumPy array).
        x_shift: Number of pixels to shift along the X-axis. Positive shifts right; negative shifts left.
        y_shift: Number of pixels to shift along the Y-axis. Positive shifts down; negative shifts up.

    Returns:
        Translated image.
    """
    # Get the image dimensions
    height, width = image.shape[:2]

    # Define the translation matrix
    translation_matrix = np.float32([[1, 0, x_shift], [0, 1, y_shift]])

    # Apply the translation using cv2.warpAffine
    translated_image = cv2.warpAffine(image, translation_matrix, (width, height), borderMode=cv2.BORDER_CONSTANT, borderValue=(0, 0, 0))

    return translated_image
```

3. Load the input image

```python
image_path = "/Girl_with_a_Pearl_Earring.jpg"  # Replace with your image path
image = cv2.imread(image_path)
```

4. Translate the image

```python
x_shift = 100  # Positive shifts right; negative shifts left
y_shift = -50  # Positive shifts down; negative shifts up
translated_image = translate_image(image, x_shift, y_shift)
```

5. Display the original and translated images

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title(f"Translated Image (X={x_shift}, Y={y_shift})")
plt.axis("off")

plt.tight_layout()
plt.show()
```

6. Save the translated image

```python
cv2.imwrite("translated_image.jpg", translated_image)
```
