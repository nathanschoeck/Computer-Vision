1. Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

2. Mirror Function

```python
def mirror_image(image, mode="horizontal"):
    """
    Mirrors (flips) an image.

    Args:
        image: Input image (NumPy array).
        mode: The flipping mode:
            - "horizontal": Flip left-to-right.
            - "vertical": Flip top-to-bottom.
            - "both": Flip both horizontally and vertically.

    Returns:
        Mirrored image.
    """
    if mode == "horizontal":
        mirrored_image = cv2.flip(image, 1)  # Flip horizontally
    elif mode == "vertical":
        mirrored_image = cv2.flip(image, 0)  # Flip vertically
    elif mode == "both":
        mirrored_image = cv2.flip(image, -1)  # Flip both axes
    else:
        raise ValueError("Invalid mode. Choose 'horizontal', 'vertical', or 'both'.")

    return mirrored_image
```

3. Load the input image

```python
image_path = "/Girl_with_a_Pearl_Earring.jpg"  # Replace with your image path
image = cv2.imread(image_path)
```

4. Apply mirroring

```python
horizontal_mirror = mirror_image(image, mode="horizontal")
vertical_mirror = mirror_image(image, mode="vertical")
both_mirror = mirror_image(image, mode="both")
```

5. Display the results

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(horizontal_mirror, cv2.COLOR_BGR2RGB))
plt.title("Horizontal Mirror")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(vertical_mirror, cv2.COLOR_BGR2RGB))
plt.title("Vertical Mirror")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(both_mirror, cv2.COLOR_BGR2RGB))
plt.title("Both Axes Mirror")
plt.axis("off")

plt.tight_layout()
plt.show()
```

6. Save the results

```python
cv2.imwrite("horizontal_mirror.jpg", horizontal_mirror)
cv2.imwrite("vertical_mirror.jpg", vertical_mirror)
cv2.imwrite("both_mirror.jpg", both_mirror)
```
