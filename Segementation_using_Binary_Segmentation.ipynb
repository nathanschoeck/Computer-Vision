Binary segmentation in computer vision is the process of dividing an image into two regions: foreground (region of interest) and background. Here's an example of performing binary segmentation in Python using OpenCV and thresholding techniques.

1. Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

2. Load the Image

```python
image_path = "/apple.jpg"  # Replace with the path to your image
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load as grayscale
```

3. Display original image

```python
plt.figure(figsize=(8, 8))
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

4. Apply binary thresholding

```python
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(8, 8))
plt.title("Binary Segmentation (Thresholding)")
plt.imshow(binary, cmap="gray")
plt.axis("off")
plt.show()
```
