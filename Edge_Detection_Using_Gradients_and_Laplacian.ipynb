Edge detection using gradients involves identifying regions in an image where there is a sharp change in intensity. This can be achieved using methods like the Sobel filter or the Laplacian filter in Python with OpenCV.

1. Import Libaries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

2. Load the image in grayscale

```python
image_path = "/Girl_with_a_Pearl_Earring.jpg"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

3. Apply Sobel filter to detect gradients

```python
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in X direction
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in Y direction
```

4. Compute the magnitude of the gradient

```python
gradient_magnitude = cv2.magnitude(sobel_x, sobel_y)
```

5. Normalize the magnitude to 0-255 for visualization

```python
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)
```

6. Compute Laplacian for edge detection

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
```

7. Display the results

```python
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.axis("off")

# Sobel X
plt.subplot(2, 2, 2)
plt.imshow(np.abs(sobel_x), cmap="gray")
plt.title("Sobel X")
plt.axis("off")

# Sobel Y
plt.subplot(2, 2, 3)
plt.imshow(np.abs(sobel_y), cmap="gray")
plt.title("Sobel Y")
plt.axis("off")

# Gradient Magnitude
plt.subplot(2, 2, 4)
plt.imshow(gradient_magnitude, cmap="gray")
plt.title("Gradient Magnitude")
plt.axis("off")

plt.tight_layout()
plt.show()

# Display Laplacian
plt.figure(figsize=(6, 6))
plt.imshow(np.abs(laplacian), cmap="gray")
plt.title("Laplacian Edge Detection")
plt.axis("off")
plt.show()
```
