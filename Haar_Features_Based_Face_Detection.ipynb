Below is Python code to perform face detection using Haar features with OpenCV. The code uses a pre-trained Haar cascade classifier provided by OpenCV to detect faces in an image.

1. Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

2. Detect faces function

```python
def detect_faces(image_path, cascade_path="haarcascade_frontalface_default.xml"):
    """
    Detects faces in an image using Haar cascade.

    Args:
        image_path: Path to the input image.
        cascade_path: Path to the Haar cascade XML file (default: frontal face).

    Returns:
        image_with_faces: Image with rectangles drawn around detected faces.
        faces: List of bounding box coordinates for the detected faces.
    """
    # Load the Haar cascade
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + cascade_path)

    # Load the image
    image = cv2.imread(image_path)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray_image,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
        flags=cv2.CASCADE_SCALE_IMAGE
    )

    # Draw rectangles around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

    return image, faces
```

3. Path to the input image

```python
image_path = "/women.jpg"
```

4. Detect faces

```python
output_image, detected_faces = detect_faces(image_path)
```

5. Display the results

```python
plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.title(f"Detected Faces: {len(detected_faces)}")
plt.axis("off")
plt.show()
```

6. Save the output image

```python
cv2.imwrite("faces_detected.jpg", output_image)
```
